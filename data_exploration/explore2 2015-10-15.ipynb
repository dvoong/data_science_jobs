```python
import re
import pprint
pp = pprint.PrettyPrinter()

class SalaryParser:
    @staticmethod
    def parse(salary):
        try:
            return int(salary)
        except Exception as e:
            pass
        
        try:
            regex = "(?P<min>[\d,.]+)\s+-\s+(?P<max>[\d,.]+)"
            m = re.match(regex, salary)
            salary = {
                "min": int(float(m.group('min').replace(",", ''))),
                "max": int(float(m.group('max').replace(",", ''))),
                "salary": None,
            }
            return salary
        except:
            pass
        
        try:
            regex = u"£(?P<min>[\d,.]+)\s+-\s+£(?P<max>[\d,.]+)"
            m = re.match(regex, salary)
            salary = {
                "min": int(float(m.group('min').replace(",", ''))),
                "max": int(float(m.group('max').replace(",", ''))),
                "salary": None,
            }
            return salary
        except:
            pass
        
        try:
            regex = u"£(?P<min>[\d,.]+)\s+to\s+£(?P<max>[\d,.]+)"
            m = re.match(regex, salary)
            salary = {
                "min": int(float(m.group('min').replace(",", ''))),
                "max": int(float(m.group('max').replace(",", ''))),
                "salary": None,
            }
            return salary
        except:
            pass
        
        try:
            regex = "[A-z ]+(?P<min>\d+)k\s+:\s+(?P<max>\d+)k"
            m = re.match(regex, salary)
            salary = {
                "min": int(m.group('min')) * 1000,
                "max": int(m.group('max')) * 1000,
                "salary": None,
            }
            return salary
        except:
            pass
        
        try:
            regex = u"[A-z ]*£(?P<min>\d+)k\s+to\s+£(?P<max>\d+)k"
            m = re.match(regex, salary)
            salary = {
                "min": int(m.group('min')) * 1000,
                "max": int(m.group('max')) * 1000,
                "salary": None,
            }
            return salary
        except:
            pass
    
        raise Exception(u"Could not parse salary.")
```

```python
successes = []
failures = []

for j in jobs:
    try:
        salary = SalaryParser.parse(j.salary)
        successes.append({
                "salary_str": j.salary,
                "salary": salary,
            })
    except Exception as e:
        failures.append({
                "salary_str": j.salary,
                "error": e.__str__(),
            })

print "n_jobs: {}".format(len(jobs))
print "n_successes: {}".format(len(successes))
print "n_failures: {}".format(len(failures))
```

```python
pp.pprint(successes)
```
